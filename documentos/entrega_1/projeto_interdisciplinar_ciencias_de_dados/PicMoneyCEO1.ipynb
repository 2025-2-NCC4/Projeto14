```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('pic.csv', sep=';') # Renomear para pic para facilitar
df.head()
```

```python
df.drop(['modelo_celular'], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df = df[df["possui_app_picmoney"] == "Sim"]
print(df)
```

```python
df.head(20)
```

```python
df.drop("possui_app_picmoney", axis=1, inplace=True)
df.head(5)
```

```python
df.drop("celular", axis=1, inplace=True)
df.head(5)
```

```python
df.describe()
```

```python
data = df["tipo_celular"].value_counts()
colors = sns.color_palette("pastel")[0:len(data)]

colors = sns.color_palette("pastel")[0:len(data)]

# Gráfico de pizza
plt.pie(data, labels=data.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Distribuição de Smartphones")
plt.show()

plt.savefig('distribuicao-aparelhos.png')
```

```python
import matplotlib.pyplot as plt

df["faixa_etaria"] = pd.cut(df["idade"], bins=[0,18,25,35,45,60,100],
                            labels=["0-18","19-25","26-35","36-45","46-60","60+"])

sns.countplot(x="faixa_etaria", data=df, palette="pastel")
plt.title("Clientes por Faixa Etária")
plt.xlabel("Faixa Etária")
plt.ylabel("Quantidade")
plt.show()

plt.savefig('faixa-etaria-clientes.png')
```

```python
sns.boxplot(x="tipo_celular", y="idade", data=df, palette="Set2")
plt.title("Idade por Tipo de Celular")
plt.show()
```

```python
sns.boxplot(x=df["idade"], color="lightgreen")
plt.title("Boxplot de Idades")
plt.show()
```

```python
sns.histplot(df["idade"], bins=10, kde=True, color="skyblue")
plt.title("Distribuição de Idades")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()

plt.savefig('distribuicao-idades.png')
```

```python
sns.countplot(x="ultimo_tipo_cupom", hue="tipo_celular", data=df, palette="Set2")
plt.title("Cupom mais usado por Tipo de Celular")
plt.show()


plt.savefig('cupons-mais-usado-por-plataforma.png')
```

```python
sns.scatterplot(x="idade", y="ultimo_valor_capturado", hue="tipo_celular", data=df)
plt.title("Idade vs Valor Capturado")
plt.show()
plt.savefig('distribuicao-idades.png')
```
