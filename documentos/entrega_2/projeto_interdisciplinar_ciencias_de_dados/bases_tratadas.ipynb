```python
from os import F_LOCK
import pandas as pd
from datetime import date

df_player = pd.read_csv('base_players.csv', sep=';')
df_transacao = pd.read_csv('cupons_capturados.csv', sep =";")
df_loja = pd.read_csv('teste_em_massa.csv', sep =";")
df_simulacao = pd.read_csv('base_paulista.csv', sep =";")
```

**Tratamento de `df_players`**


*   Ajuste das datas
*   Checagem se as idades são fidedignas



```python
df = df_player

df["data_nascimento"] = pd.to_datetime(df["data_nascimento"], dayfirst=True, errors="coerce")

teste = pd.Timestamp(date.today())

df["idade"] = df["data_nascimento"].apply(
    lambda d: teste.year - d.year - ((teste.month, teste.day) < (d.month, d.day)) if pd.notnull(d) else None
)


df.to_csv(r"base_players-limpo.csv", index=False, encoding="utf-8-sig")
```
